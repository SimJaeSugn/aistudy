### ν•¨μ μ •μμ™€ ν™μ©


````python
def format_price(price):
    """κ°€κ²©μ„ μ½κΈ° μ‰¬μ΄ ν•νƒλ΅ ν¬λ§·ν…ν•λ” ν•¨μ"""
    if price >= 100000000:  # 1μ–µ μ΄μƒ
        return f"{price/100000000:.1f}μ–µμ›"
    elif price >= 10000:    # 1λ§ μ΄μƒ
        return f"{price/10000:.1f}λ§μ›"
    else:
        return f"{price:,}μ›"

def calculate_change_rate(old_price, new_price):
    """κ°€κ²© λ³€λ™λ¥ μ„ κ³„μ‚°ν•λ” ν•¨μ"""
    change_rate = ((new_price - old_price) / old_price) * 100
    return round(change_rate, 2)
````

````python
# ν•¨μ μ‚¬μ© μμ‹
prices = [50000000, 1500000, 25000, 500]

print("κ°€κ²© ν¬λ§·ν… μμ‹:")
for price in prices:
    formatted = format_price(price)
    print(f"{price:,}μ› β†’ {formatted}")

print("\nλ³€λ™λ¥  κ³„μ‚° μμ‹:")
old_price = 50000000
new_price = 52500000
change_rate = calculate_change_rate(old_price, new_price)
print(f"{format_price(old_price)} β†’ {format_price(new_price)} ({change_rate:+.2f}%)")
````

### μ°¨νΈ λ°μ΄ν„° μ΅°ν


````python
# λΉ„νΈμ½”μΈ μΌλ΄‰ λ°μ΄ν„° μ΅°ν (μµκ·Ό 7μΌ)
url = "https://api.upbit.com/v1/candles/days"
headers = {"accept": "application/json"}
params = {
    "market": "KRW-BTC",
    "count": 7
}
response = requests.get(url, headers=headers, params=params)
candle_data = response.json()

# λ¦¬μ¤νΈμ™€ λ”•μ…”λ„λ¦¬λ΅ λ°μ΄ν„° λ³€ν™
btc_daily = []
for candle in reversed(candle_data):  # μ‹κ°„μμΌλ΅ μ •λ ¬
    date_str = candle['candle_date_time_kst'].replace('T', ' ')
    date_obj = datetime.fromisoformat(date_str)

    btc_daily.append({
        'date': date_obj.strftime('%Y-%m-%d %H:%M:%S'),
        'open': candle['opening_price'],
        'high': candle['high_price'],
        'low': candle['low_price'],
        'close': candle['trade_price'],
        'volume': candle['candle_acc_trade_volume'],
        'value': candle['candle_acc_trade_price']
    })

# λ°μ΄ν„° μ¶λ ¥
print("λΉ„νΈμ½”μΈ μµκ·Ό 7μΌ μΌλ΄‰ λ°μ΄ν„°:")
print("-" * 80)
print(f"{'λ‚ μ§':>20} {'μ‹κ°€':>12} {'κ³ κ°€':>12} {'μ €κ°€':>12} {'μΆ…κ°€':>12} {'κ±°λλ‰':>10}")
print("-" * 80)

for day_data in btc_daily:
    date = day_data['date'][:10]  # λ‚ μ§λ§ ν‘μ‹
    open_price = f"{day_data['open']:,.0f}"
    high_price = f"{day_data['high']:,.0f}"
    low_price = f"{day_data['low']:,.0f}"
    close_price = f"{day_data['close']:,.0f}"
    volume = f"{day_data['volume']:.1f}"

    print(f"{date:>20} {open_price:>12} {high_price:>12} {low_price:>12} {close_price:>12} {volume:>10}")

print(f"\nλ°μ΄ν„° νƒ€μ…: {type(btc_daily)}")
print(f"λ°μ΄ν„° κ°μ: {len(btc_daily)}κ°")
print(f"μ²« λ²μ§Έ λ°μ΄ν„° ν‚¤: {list(btc_daily[0].keys())}")

# νΉμ • λ‚ μ§ λ°μ΄ν„° μ ‘κ·Ό μμ‹
print(f"\nμ²« λ²μ§Έ λ‚ μ§ λ°μ΄ν„°:")
first_day = btc_daily[0]
for key, value in first_day.items():
    if key == 'date':
        print(f"{key}: {value}")
    elif key in ['volume']:
        print(f"{key}: {value:.2f}")
    else:
        print(f"{key}: {value:,.0f}")
````

````python
# λ¬Έμ  1μ λ‹µμ•μ„ μ‘μ„±ν•΄μ£Όμ„Έμ”
import requests
from datetime import datetime
import time
import json
from pprint import pprint
import time

# 1. λ³΄μ  μ•”νΈν™”νμ™€ μλ‰μ„ λ”•μ…”λ„λ¦¬λ΅ μ •μ 
portfolio = {
    "KRW-BTC":0.1,
    "KRW-ETH":3,
    "KRW-XRP":1000
}

# 2. κ° μ•”νΈν™”νμ ν„μ¬ κ°€μΉ κ³„μ‚° 
def get_current_prices_api(portfolio):
    """ν„μ¬κ°€ μ΅°ν"""
    tickers = [ticker for ticker,value in portfolio.items()]
    url = "https://api.upbit.com/v1/ticker"
    headers = {"accept":"application/json"}
    params = {"markets":",".join(tickers)}  

    response = requests.get(url,headers=headers,params=params)    
    prices_data = response.json()

    prices = {}
    for data in prices_data:
        prices[data["market"]] = data["trade_price"]
    
    return prices

# 3. μ „μ²΄ ν¬νΈν΄λ¦¬μ¤ κ°€μΉ κ³„μ‚° 
def analyze_portfolio(portfolio):
    
    # κ° μ•”νΈν™”νμ ν„μ¬ κ°€μΉ κ³„μ‚° 
    prices = get_current_prices_api(portfolio)

    portfolio_analysis = {}
    total_value = 0
    worth = 0
    print(portfolio)
    # κ° μ•”νΈν™”νλ³„ λ¶„μ„
    for tickers,cnt in portfolio.items():        
        worth = prices[tickers] * cnt
        portfolio_analysis[tickers] = worth
        total_value += worth

    # 4. κ° μ•”νΈν™”νμ ν¬νΈν΄λ¦¬μ¤ λ‚΄ λΉ„μ¤‘ κ³„μ‚° 
    weight_analysis = {}    
    for tickers,cnt in portfolio.items():
       weight = (portfolio_analysis[tickers] / total_value) * 100
       weight_analysis[tickers] = weight

    

    # λ²μ„κ²°κ³Ό μ¶λ ¥
    print("-" * 82)
    print(f"{'μ½”μΈ':>18} {'μλ‰':>10} {'ν„μ¬κ°€':>12} {'κ°€μΉ':>13} {'λΉ„μ¤‘':>13}")
    print("-" * 82)    
    for coin,amount in portfolio.items():
        worth = portfolio_analysis[coin]
        now_price = prices[coin]
        weight = weight_analysis[coin]
        print(f"{coin:>20} {amount:>12} {now_price:>15,} {worth:>15,} {weight:>16.2f}")


analyze_portfolio(portfolio)
````

````python
# λ¬Έμ  2μ λ‹µμ•μ„ μ‘μ„±ν•΄μ£Όμ„Έμ”
import requests
from datetime import datetime
import time
import json
from pprint import pprint
import time

# 1 .λ©ν‘κ°€ μ„¤μ •
goals = [
    {"ticker":"KRW-BTC","up_price":1.05,"down_price":0.95,"now_price":136149000},
    {"ticker":"KRW-ETH","up_price":1.05,"down_price":0.95,"now_price":4543000},
    {"ticker":"KRW-XRP","up_price":1.05,"down_price":0.95,"now_price":3273}
]
# ν„μ¬κ°€ μ΅°ν
def get_single_price_api(ticker):
    """ν„μ¬κ°€ λ‹¨κ±΄ μ΅°ν"""
    url = "https://api.upbit.com/v1/ticker"
    headers = {"accept":"application/json"}
    params = {"markets":ticker}
    response = requests.get(url,headers=headers,params=params)    
    btc_data = response.json()[0]    
    date_object = datetime.fromtimestamp(btc_data["timestamp"]/1000)
    date_str = date_object.strftime('%H:%M:%S')
    info = {
        "trade_price":btc_data["trade_price"],
        "read_tm":date_str
    }
    

    return info


# 3. λ¨ν‹°ν„°λ§ ν•¨μ
def price_alert_system(goals,count=10,step=5):
    """μ£Όμ–΄μ§„ μ΅°κ±΄μ„ μ μ©ν• λ¨λ‹ν„°λ§ μ§„ν–‰"""
    print("λ©ν‘ μƒν•κ°€,ν•ν•κ°€ λΉ„μ¨")
    print("-" * 50)    
    print(f"{'μ½”μΈ':>16} {'ν„μ¬κ°€':>6} {'μƒν•κ°€':>6} {'ν•ν•κ°€':>8}")
    print("-" * 50)  
    goal_details = []
    for goal_info in goals:        
        print(f"{goal_info["ticker"]:>18}{goal_info["now_price"]:>10}{goal_info["now_price"]*goal_info["up_price"]:>10}{goal_info["now_price"]*goal_info["down_price"]:>12}")
        
        goal_detail = {}
        goal_detail["ticker"] = goal_info["ticker"]
        goal_detail["up_price"] = goal_info["ticker"] = goal_info["now_price"]*goal_info["up_price"]
        goal_detail["down_price"] = goal_info["ticker"] = goal_info["now_price"]*goal_info["down_price"]   
        goal_detail["now_price"] = goal_info["now_price"]

        goal_details.append(goal_detail)

    print("λ¨λ‹ν„°λ§μ„ μ‹μ‘ν•©λ‹λ‹¤.")
    for i in range(count):
        print("="*33,f" {i+1}λ²μ§Έ μ΅°ν ","="*33)        
        print("-" * 80)    
        print(f"{'μ½”μΈ':>8} {'ν„μ¬κ°€':>12} {'λ©ν‘μƒν•κ°€':>12} {'λ©ν‘ν•ν•κ°€':>12} {'μƒνƒ':>5} {'μ‹κ°„':>8}")
        print("-" * 80)
        for coin in goal_details:        
            price_info = get_single_price_api(coin["ticker"])
            if coin["up_price"] <= price_info["trade_price"]:
                # μƒν•κ°€ λ„λ‹¬
                state = "π”Ί"
            elif coin["down_price"] >= price_info["trade_price"]:
                # ν•ν•κ°€ λ„λ‹¬
                state = "π”»"
            else:
                # μ •μƒ λ²”μ„
                state = "β…"

            print(f"{coin["ticker"]:>10} {price_info["trade_price"]:>15,} {coin["up_price"]:>17,}{coin["down_price"]:>17,}{state:>8} {price_info["read_tm"]:>10}")

        time.sleep(step)

price_alert_system(goals,10)
````

````python
# λ¬Έμ  3μ λ‹µμ•μ„ μ‘μ„±ν•΄μ£Όμ„Έμ”
import requests
from datetime import datetime,timedelta
import time
import json
from pprint import pprint
import time

def format_price(price):
    """κ°€κ²©μ„ μ½κΈ° μ‰¬μ΄ ν•νƒλ΅ ν¬λ§·ν…ν•λ” ν•¨μ"""
    if price >= 100000000:  # 1μ–µ μ΄μƒ
        return f"{price/100000000:.1f}μ–µμ›"
    elif price >= 10000:    # 1λ§ μ΄μƒ
        return f"{price/10000:.1f}λ§μ›"
    else:
        return f"{price:,.1f}μ›"
    
# 1. ν¬μ‚¬ μ‹μ κ³Ό ν¬μ κΈμ•΅ μ„¤μ •
investment_goal = [
    {"market":"KRW-BTC","target_day":365,"money":280000000},
    {"market":"KRW-ETH","target_day":365,"money":57000000},
    {"market":"KRW-XRP","target_day":365,"money":25490000},
]

# ν„μ¬κ°€ μ΅°ν
def get_single_price_api(ticker):
    """ν„μ¬κ°€ λ‹¨κ±΄ μ΅°ν"""
    url = "https://api.upbit.com/v1/ticker"
    headers = {"accept":"application/json"}
    params = {"markets":ticker}
    response = requests.get(url,headers=headers,params=params)    
    btc_data = response.json()[0]    
    date_object = datetime.fromtimestamp(btc_data["timestamp"]/1000)
    date_str = date_object.strftime('%H:%M:%S')
    info = {
        "trade_price":btc_data["trade_price"],
        "read_tm":date_str
    }
    return info

# 2. ν•΄λ‹Ή μ‹μ μ κ°€κ²©κ³Ό ν„μ¬ κ°€κ²© λΉ„κµ
def get_historical_data_api(investment_goal):
    """κ³Όκ±° λ°μ΄ν„° μ΅°ν™”"""
    url = "https://api.upbit.com/v1/candles/days"
    headers = {"accept": "application/json"}

    today = datetime.now()
    rtn_data = {}
    for coin in investment_goal:
        days_ago = timedelta(days=coin["target_day"])    
        date_20_days_ago = today - days_ago
        target_date = date_20_days_ago.strftime('%Y-%m-%d %H:%M:%S')
        params = {
            "market": coin["market"],
            "to":target_date
        }
        response = requests.get(url, headers=headers,params=params)
        candle_data = response.json()
        coin["trade_price"] = candle_data[0]["trade_price"]
        coin["target_date"] = target_date
        # ν¬μμ‹μ μ μΆ…κ°€λ°μ΄ν„°λ¥Ό μ¶”κ°€
        rtn_data[coin["market"]] = coin

    return rtn_data
    
       
# 3. μμµλ¥ κ³Ό μμ…κΈ κ³„μ‚°
def calulate_investment_return(investment_goal):
    hisotry_info = get_historical_data_api(investment_goal)
    print(hisotry_info)
    print("-" * 150)    
    print(f"{'μ½”μΈ':>8} {'ν¬μ μΌμ':>8} {'ν¬μ κΈμ•΅':>8} {'ν¬μμ‹μ κ°€κ²©':>10} {'κµ¬λ§¤ μλ‰':>8} {'ν„μ¬ κ°€κ²©':>12} {'ν„μ¬ κ°€μΉ':>8} {'μ†μµ':>13} {'μμµλ¥ ':>8}")
    print("-" * 150)
    
    for coin,data in hisotry_info.items():
        # ν¬μμ‹μ  κ°€κ²©
        amount = data["money"] / data["trade_price"]
        # ν„μ¬ κ°€μΉ
        now_info = get_single_price_api(coin)
        now_worth =  amount  * now_info["trade_price"]

        # μ†μµ κ°€μΉ κ³„μ‚°
        loss = now_worth - data["money"]
        rate = (loss / data["money"]) * 100

        print(f"{coin:>10} {data["target_date"][:10]:>12} {format_price(data["money"]):>10}{format_price(data["trade_price"]):>15}{amount:>12.2f}{format_price(now_info["trade_price"]):>17}{format_price(now_worth):>10}{format_price(loss):>15}{rate:>12.2f}%")

calulate_investment_return(investment_goal)
````
