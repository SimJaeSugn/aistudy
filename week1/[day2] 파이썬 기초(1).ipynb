# 변수와 자료형

## 변수란 무엇인가?

변수는 **데이터를 저장하는 상자**라고 생각할 수 있습니다.
즉, 어떤 값을 임시로 저장해두고 필요할 때마다 꺼내 쓸 수 있도록 하는 역할을 합니다.
프로그래밍에서 변수는 메모리의 특정 공간에 이름을 붙여서, 그 공간에 데이터를 저장하거나 읽어올 수 있게 해줍니다.

- 상자에 이름표를 붙이는 것을 '변수명'이라고 하며, 이 이름을 통해 저장된 값을 구분하고 사용할 수 있습니다.
- 상자 안에 값을 넣는 과정을 '할당'이라고 합니다. 예를 들어, `a = 10`이라고 하면 변수 a에 10이라는 값을 저장하는 것입니다.
- 필요할 때 변수명을 사용하여 상자 안의 값을 꺼내 쓸 수 있는데, 이를 '참조'라고 합니다.

### 변수 사용의 장점:
1. **재사용성**: 한 번 저장한 값을 여러 번 사용할 수 있습니다. 예를 들어, 같은 값을 여러 곳에서 반복해서 사용할 때 변수에 저장해두면 편리합니다.
2. **가독성**: 변수명을 통해 코드의 의미를 쉽게 파악할 수 있어, 코드를 읽고 이해하기 쉬워집니다. 예를 들어, 단순히 숫자 100을 사용하는 것보다 `score = 100`처럼 변수명을 사용하면 코드의 목적이 명확해집니다.
3. **유지보수**: 코드에서 사용되는 값을 한 곳(변수)에서만 변경하면 되므로, 코드 전체를 일일이 수정할 필요가 없어 관리가 쉽습니다. 예를 들어, 변수에 저장된 값을 바꾸면 그 변수를 사용하는 모든 부분에 자동으로 적용됩니다.


```python
# 변수 만들기와 사용하기

# 1. 변수에 값 저장하기 (할당)
name = "김철수"
age = 25
height = 175.5

# 2. 변수 값 출력하기
print("이름:", name)
print("나이:", age)
print("키:", height)

# 3. 변수 값 변경하기
age = 26  # 생일이 지나서 나이가 늘었다고 가정
print("변경된 나이:", age)

# 4. 변수를 이용한 계산
birth_year = 2024 - age
print("출생년도:", birth_year)
```

## 기본 자료형 (Data Types)

파이썬에서 데이터를 다룰 때 가장 기본적으로 사용되는 자료형에는 여러 가지가 있습니다. 각각의 자료형은 저장할 수 있는 값의 종류와 그 값에 대해 할 수 있는 연산이 다릅니다.

### 1. 숫자형 (Numbers)
- **정수 (int)**: 소수점이 없는 숫자를 의미합니다. 예를 들어 1, 100, -5와 같이 양수, 음수, 0 모두 정수형에 해당합니다. 정수형은 사칙연산(+, -, *, /, //, %, **) 등 다양한 연산에 사용할 수 있습니다.
- **실수 (float)**: 소수점이 있는 숫자를 의미합니다. 예를 들어 3.14, -2.5, 0.0 등이 있습니다. 실수형은 정수형과 마찬가지로 사칙연산이 가능하며, 소수점 이하의 값을 표현할 수 있습니다. 실수형은 주로 측정값, 비율, 평균 등 소수점이 필요한 데이터를 다룰 때 사용합니다.

### 2. 문자열 (String)
- **문자열 (str)**: 문자, 단어, 문장 등 문자들의 집합을 저장하는 자료형입니다. 예를 들어 "안녕하세요", "Python", "123" 등은 모두 문자열입니다. 문자열은 작은따옴표(' ')나 큰따옴표(" ")로 감싸서 표현합니다.
- 문자열은 인덱싱(문자 하나씩 접근), 슬라이싱(일부만 추출), 덧셈(문자열 이어붙이기), 곱셈(반복) 등 다양한 연산이 가능합니다. 또한 문자열의 길이를 구하거나, 특정 문자를 찾는 등 여러 가지 내장 함수를 사용할 수 있습니다.

### 3. 불린 (Boolean)
- **불린 (bool)**: 참(True) 또는 거짓(False) 두 가지 값만 가질 수 있는 자료형입니다. 주로 조건문(if, while 등)에서 조건의 결과를 표현할 때 사용합니다.
- 불린형은 비교 연산(==, !=, >, <, >=, <=)이나 논리 연산(and, or, not)에서 결과로 자주 사용되며, 프로그램의 흐름을 제어하는 데 매우 중요합니다.

이 외에도 파이썬에는 리스트(list), 튜플(tuple), 딕셔너리(dict), 집합(set)과 같은 다양한 자료형이 있지만, 위의 세 가지가 가장 기본적인 자료형입니다.


```python
# 자료형 예시와 type() 함수 사용하기

# 1. 정수형 (int)
student_count = 30
temperature = -5
print("학생 수:", student_count, "| 자료형:", type(student_count))
print("온도:", temperature, "| 자료형:", type(temperature))

print()  # 빈 줄 출력

# 2. 실수형 (float)
pi = 3.14159
weight = 65.5
print("원주율:", pi, "| 자료형:", type(pi))
print("몸무게:", weight, "| 자료형:", type(weight))


# 3. 문자열 (str)
greeting = "안녕하세요!"
course_name = 'Python 기초'
print("인사말:", greeting, "| 자료형:", type(greeting))
print("과정명:", course_name, "| 자료형:", type(course_name))


# 4. 불린형 (bool)
is_student = True
is_raining = False
print("학생인가요?", is_student, "| 자료형:", type(is_student))
print("비가 오나요?", is_raining, "| 자료형:", type(is_raining))
```

## 변수 이름 짓기 규칙

변수는 데이터를 저장하고 관리하는 데 사용되므로, 변수명을 잘 짓는 것은 코드의 가독성과 유지보수에 매우 큰 영향을 미칩니다. 변수 이름을 명확하고 일관성 있게 작성하면, 코드를 읽는 사람이 변수의 역할을 쉽게 이해할 수 있습니다. 아래는 변수 이름을 지을 때 반드시 지켜야 하는 규칙과, 더 좋은 코드를 위해 권장되는 사항들입니다.

### 규칙 (반드시 지켜야 함)
1. **영문자, 숫자, 밑줄(_)만 사용** 가능  
   - 변수 이름에는 한글, 특수문자(공백, !, @ 등)를 사용할 수 없습니다.  
   - 예시: `score1`, `user_name`, `total_3`
2. **숫자로 시작할 수 없음**  
   - 변수 이름의 첫 글자는 반드시 영문자 또는 밑줄(_)이어야 하며, 숫자로 시작하면 오류가 발생합니다.  
   - 예시: `1st_student` (X), `student1` (O)
3. **대소문자를 구분함**  
   - 파이썬에서는 대문자와 소문자를 다르게 인식합니다.  
   - 예시: `name`과 `Name`은 서로 다른 변수입니다.
4. **예약어(키워드) 사용 불가**  
   - 파이썬에서 이미 특별한 의미로 사용되는 단어(예약어)는 변수명으로 사용할 수 없습니다.  
   - 예시: `if`, `for`, `while`, `class`, `def` 등은 변수명으로 사용할 수 없습니다.

### 권장사항 (더 좋은 코드를 위해)
1. **의미 있는 이름 사용**  
   - 변수의 역할이 드러나도록 이름을 짓는 것이 좋습니다.  
   - 예시: `a`보다는 `student_age`, `total_score`처럼 구체적인 이름을 사용합니다.
2. **스네이크 케이스(snake_case) 사용**  
   - 여러 단어로 이루어진 변수명은 단어 사이를 밑줄(_)로 구분하는 스네이크 케이스를 사용하는 것이 일반적입니다.  
   - 예시: `student_name`, `average_score`
3. **너무 길지 않게**  
   - 변수명이 너무 길면 오히려 읽기 어렵고, 너무 짧으면 의미를 알기 어렵습니다.  
   - 적당한 길이로 변수명을 작성하는 것이 좋습니다.
4. **일관성 유지**  
   - 한 프로젝트 내에서는 변수명 스타일을 통일해서 사용해야 합니다.  
   - 예를 들어, 모두 스네이크 케이스로 작성하거나, 모두 소문자로 작성하는 등 일관성을 지키는 것이 중요합니다.

이러한 규칙과 권장사항을 지키면, 코드의 가독성이 높아지고 협업 시에도 의사소통이 쉬워집니다.


```python
# 변수명 예시 - 좋은 예와 나쁜 예

# 좋은 변수명 예시
student_name = "김영희"
student_age = 20
total_score = 95
is_passed = True

print("=== 좋은 변수명 사용 ===")
print(f"학생 이름: {student_name}")
print(f"나이: {student_age}")
print(f"총점: {total_score}")
print(f"합격 여부: {is_passed}")

print()

# 나쁜 변수명 예시 (동작은 하지만 권장하지 않음)
n = "김영희"  # 너무 짧고 의미 불분명
a = 20       # 무엇을 의미하는지 알 수 없음
s = 95       # 점수인지 다른 값인지 모호
p = True     # 무엇이 참인지 알 수 없음

print("=== 나쁜 변수명 사용 ===")
print(f"이름: {n}")
print(f"나이: {a}")
print(f"점수: {s}")
print(f"통과: {p}")

# 같은 기능이지만 어떤 코드가 더 이해하기 쉬운가요?
```

# 기본 입출력

## print() 함수 활용하기

`print()` 함수는 화면에 내용을 출력하는 가장 기본적인 함수입니다.

### print() 함수의 다양한 사용법:
1. **기본 출력**: `print("내용")`
2. **여러 값 출력**: `print(값1, 값2, 값3)`
3. **구분자 변경**: `print(값1, 값2, sep="구분자")`
4. **줄바꿈 제어**: `print("내용", end="")`


```python
# print() 함수의 다양한 사용법

# 1. 기본 출력
print("안녕하세요!")
print("파이썬을 배우고 있습니다.")

print("=" * 30)  # 구분선

# 2. 여러 값을 한 번에 출력 (기본적으로 공백으로 구분)
name = "김철수"
age = 25
city = "서울"
print("이름:", name, "나이:", age, "거주지:", city)

print("=" * 30)

# 3. 구분자(separator) 변경하기
print("사과", "바나나", "딸기", sep=" | ")
print("2024", "12", "25", sep="-")
print("010", "1234", "5678", sep="-")

print("=" * 30)

# 4. 줄바꿈 제어하기 (end 매개변수)
print("첫 번째 줄", end=" ")
print("같은 줄에 출력됩니다!")

print("로딩 중", end="")
for i in range(5):
    print(".", end="")
print(" 완료!")

print("=" * 30)

# 5. 여러 줄 문자열 출력
message = """
안녕하세요!
파이썬 기초 강의에 오신 것을 환영합니다.
함께 즐겁게 학습해봅시다!
"""
print(message)
```

## input() 함수로 사용자 입력 받기

`input()` 함수는 사용자로부터 키보드 입력을 받는 함수입니다.

### 중요한 특징:
- **항상 문자열(str)로 반환**: 숫자를 입력해도 문자열로 저장됨
- **프롬프트 메시지 표시 가능**: `input("메시지")`
- **Enter 키를 누를 때까지 대기**: 사용자 입력을 기다림

## 문자열 포매팅 (f-string)

문자열 포매팅은 변수의 값을 문자열에 삽입하는 방법입니다.
파이썬 3.6부터 도입된 **f-string**이 가장 현대적이고 편리한 방법입니다.

### f-string 사용법:
1. 문자열 앞에 `f` 또는 `F` 붙이기
2. 중괄호 `{}` 안에 변수나 표현식 넣기
3. 자동으로 값이 문자열로 변환되어 삽입됨

### 장점:
- **가독성이 좋음**: 어떤 값이 어디에 들어가는지 명확
- **성능이 빠름**: 다른 포매팅 방법보다 빠름
- **표현식 사용 가능**: 계산식도 직접 넣을 수 있음


```python
# f-string 사용 예시

# 변수 정의
name = "김민수"
age = 22
height = 175.5
weight = 70.2

print("=== 기본 f-string 사용 ===")
print(f"이름: {name}")
print(f"나이: {age}세")
print(f"키: {height}cm")
print(f"몸무게: {weight}kg")


# 표현식 사용 (계산식을 직접 넣기)
print("=== 표현식 사용 ===")
print(f"내년 나이: {age + 1}세")
print(f"BMI 지수: {weight / (height/100)**2:.2f}")  # 소수점 2자리까지
print(f"키(미터): {height/100:.1f}m")  # 소수점 1자리까지


# 다양한 포매팅 옵션
print("=== 포매팅 옵션 ===")
price = 1234567
print(f"가격: {price:,}원")  # 천 단위 구분자
print(f"가격: {price:>10,}원")  # 오른쪽 정렬, 10자리
print(f"가격: {price:<10,}원")  # 왼쪽 정렬, 10자리
print(f"가격: {price:^15,}원")  # 가운데 정렬, 15자리


# 퍼센트와 진법 표현
print("=== 특수 포매팅 ===")
ratio = 0.8567
print(f"성공률: {ratio:.1%}")  # 퍼센트로 표현
print(f"성공률: {ratio:.2%}")  # 소수점 2자리 퍼센트

number = 255
print(f"10진수: {number}")
print(f"16진수: {number:x}")  # 소문자 16진수
print(f"16진수: {number:X}")  # 대문자 16진수
print(f"2진수: {number:b}")   # 2진수


# 문자열 조합
print("=== 복합 정보 출력 ===")
subject = "파이썬"
score = 95
grade = "A"

print(f"{name}님의 {subject} 성적은 {score}점({grade})입니다.")
print(f"평균보다 {score - 80}점 높습니다!")

# 여러 줄 f-string
info = f"""
학생 정보 카드
================
이름: {name}
나이: {age}세
키: {height}cm
몸무게: {weight}kg
BMI: {weight / (height/100)**2:.1f}
"""
print(info)
```

## 문자열 템플릿과 .format() 메서드

f-string 외에도 파이썬에서는 **문자열 템플릿**을 만들어 재사용할 수 있는 방법이 있습니다.
`.format()` 메서드를 사용하면 미리 정의된 템플릿에 나중에 값을 채워넣을 수 있습니다.

### .format() 메서드의 특징:
1. **재사용 가능한 템플릿**: 한 번 정의하고 여러 번 사용
2. **위치 인수**: `{0}`, `{1}` 형태로 순서대로 값 삽입
3. **키워드 인수**: `{name}`, `{age}` 형태로 이름으로 값 삽입
4. **혼합 사용**: 위치 인수와 키워드 인수를 함께 사용 가능

### 언제 사용하면 좋을까?
- 같은 형식의 메시지를 반복적으로 만들 때
- 템플릿을 미리 정의하고 나중에 값을 채울 때
- 설정 파일이나 메시지 템플릿을 관리할 때


```python
# .format() 메서드 사용 예시

print("=== 기본 .format() 사용법 ===")

# 1. 위치 인수 사용 (순서대로)
template1 = "안녕하세요! 저는 {}이고, {}세입니다."
message1 = template1.format("김철수", 25)
print(message1)

# 인덱스를 명시적으로 지정
template2 = "안녕하세요! 저는 {0}이고, {1}세입니다."
message2 = template2.format("이영희", 23)
print(message2)


# 2. 키워드 인수 사용 (이름으로)
print("=== 키워드 인수 사용 ===")
template3 = "안녕하세요! 저는 {name}이고, {age}세입니다."
message3 = template3.format(name="박민수", age=27)
print(message3)

# 같은 템플릿으로 다른 값들 사용
message4 = template3.format(name="최수진", age=24)
print(message4)


# 3. 사용자가 제시한 예시 - 역할 기반 템플릿
print("=== 역할 기반 템플릿 예시 ===")
template = """
당신은 {role}입니다.
다음 주제에 대해 {style} 스타일로 설명해주세요:
주제: {topic}
"""

# 템플릿 사용 예시들
message_teacher = template.format(
    role="선생님",
    style="쉽고 친근한",
    topic="파이썬 변수"
)
print("교사용 메시지:")
print(message_teacher)

message_expert = template.format(
    role="전문가",
    style="전문적이고 상세한",
    topic="데이터 구조"
)
print("전문가용 메시지:")
print(message_expert)
```

## 문자열 템플릿과 %s 포매팅

`%s` 포매팅은 오래전부터 사용되어 온 방식으로, 문자열 안에 `%` 기호와 포맷 코드를 사용해 값을 넣을 수 있습니다.

`.format()`이나 f-string이 등장하기 전에는 `%s`, `%d`, `%f`와 같은 포맷 코드를 많이 사용했습니다. 이 방식은 C언어의 printf와 비슷한 느낌으로, 문자열을 미리 만들어두고 나중에 값을 채워넣을 수 있습니다.

### %s 포매팅의 특징
1. **간단한 문법**: 문자열 안에 `%s`를 넣고, 뒤에 `%` 연산자로 값을 전달합니다.
2. **여러 값 삽입 가능**: 값이 여러 개일 때는 튜플로 묶어서 전달합니다.
3. **다양한 타입 지원**: `%s`는 어떤 타입이든 문자열로 변환해서 출력합니다. 숫자에는 `%d`, 실수에는 `%f`를 사용할 수 있습니다.
4. **재사용 가능**: 템플릿을 미리 만들어두고 여러 번 사용할 수 있습니다.

### 언제 사용하면 좋을까?
- 간단한 문자열 포매팅이 필요할 때
- 기존 코드나 예전 파이썬 코드와 호환이 필요할 때
- 빠르게 문자열을 조합해서 출력하고 싶을 때


```python

### %s 포매팅 코드 실습

# 1. 기본 사용법
name = "홍길동"
age = 30
print("안녕하세요, 제 이름은 %s입니다. 나이는 %s세입니다." % (name, age))


```

## f-string vs .format() 비교

두 방법 모두 유용하지만, 상황에 따라 적합한 방법이 다릅니다.

### f-string의 장점:
- **간결하고 읽기 쉬움**: 변수가 바로 보임
- **성능이 빠름**: 런타임에서 더 효율적
- **표현식 사용 가능**: `{age + 1}` 같은 계산 직접 가능

### .format()의 장점:
- **템플릿 재사용**: 한 번 정의하고 여러 번 사용
- **동적 생성**: 템플릿을 파일에서 읽거나 변수로 저장 가능
- **복잡한 포매팅**: 더 세밀한 포매팅 제어

### 선택 기준:
- **즉석에서 한 번 사용**: f-string 추천
- **반복적으로 사용할 템플릿**: .format() 추천
- **설정 파일이나 외부에서 템플릿 관리**: .format() 추천


```python
# f-string vs .format() 비교 예시

print("=== 같은 결과, 다른 방법 ===")

# 변수 정의
name = "김코더"
age = 25
language = "Python"

# 1. f-string 방식 (즉석에서 사용)
message_f = f"안녕하세요! 저는 {name}이고, {age}세입니다. {language}를 배우고 있어요!"
print("f-string:", message_f)

# 2. .format() 방식 (템플릿 재사용)
template = "안녕하세요! 저는 {name}이고, {age}세입니다. {language}를 배우고 있어요!"
message_format = template.format(name=name, age=age, language=language)
print(".format():", message_format)

print()

# 3. 템플릿 재사용의 장점
print("=== 템플릿 재사용 예시 ===")
greeting_template = "안녕하세요, {name}님! {course} 수강을 환영합니다."

# 같은 템플릿으로 여러 학생에게 인사
students = [
    {"name": "김학생", "course": "파이썬 기초"},
    {"name": "이학생", "course": "데이터 분석"},
    {"name": "박학생", "course": "웹 개발"}
]

for student in students:
    greeting = greeting_template.format(**student)
    print(greeting)

print()
```
