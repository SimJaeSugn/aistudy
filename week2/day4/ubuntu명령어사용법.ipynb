## 파일 및 디렉토리 관리


### pwd
현재 작업 디렉토리 경로 출력


```python
pwd
```

### ls
파일 및 디렉토리 목록 출력


```python
# 기본 목록
ls

# 상세 정보 포함
ls -l

# 숨김 파일 포함
ls -a

# 상세 정보 + 숨김 파일
ls -la

# 사람이 읽기 쉬운 크기 표시
ls -lh
```

### cd
디렉토리 이동


```python
# 특정 디렉토리로 이동
cd /home/user

# 홈 디렉토리로 이동
cd ~

# 이전 디렉토리로 이동
cd -

# 상위 디렉토리로 이동
cd ..
```

### mkdir
디렉토리 생성


```python
# 디렉토리 생성
mkdir mydir

# 하위 디렉토리까지 한번에 생성
mkdir -p parent/child/grandchild
```

### touch
빈 파일 생성 또는 타임스탬프 갱신


```python
touch newfile.txt
```

### cp
파일 및 디렉토리 복사


```python
# 파일 복사
cp file1.txt file2.txt

# 디렉토리 복사 (재귀적)
cp -r dir1 dir2
```

### mv
파일 및 디렉토리 이동 또는 이름 변경


```python
# 파일 이동
mv file.txt /home/user/

# 파일 이름 변경
mv oldname.txt newname.txt
```

### rm
파일 및 디렉토리 삭제


```python
# 파일 삭제
rm file.txt

# 디렉토리 삭제 (재귀적)
rm -r mydir

# 강제 삭제 (확인 없이)
rm -f file.txt

# 디렉토리 강제 삭제
rm -rf mydir
```

## 파일 내용 확인


### cat
파일 전체 내용 출력


```python
cat file.txt

# 여러 파일 동시 출력
cat file1.txt file2.txt
```

### head
파일의 앞부분 출력 (기본 10줄)


```python
head file.txt

# 처음 20줄 출력
head -n 20 file.txt
```

### tail
파일의 뒷부분 출력 (기본 10줄)


```python
tail file.txt

# 마지막 20줄 출력
tail -n 20 file.txt

# 실시간 로그 확인 (파일 변화 감시)
tail -f /var/log/syslog
```

### less
파일을 페이지 단위로 확인 (스크롤 가능)


```python
less file.txt
# 종료: q
# 검색: /keyword
```

## 파일 검색


### find
파일 및 디렉토리 검색


```python
# 현재 디렉토리에서 파일명으로 검색
find . -name "file.txt"

# 특정 확장자 검색
find . -name "*.txt"

# 디렉토리만 검색
find . -type d

# 파일만 검색
find . -type f
```

### grep
파일 내용에서 텍스트 검색


```python
# 파일에서 특정 문자열 검색
grep "keyword" file.txt

# 대소문자 구분 없이 검색
grep -i "keyword" file.txt

# 재귀적으로 모든 파일 검색
grep -r "keyword" .

# 줄 번호와 함께 출력
grep -n "keyword" file.txt
```

## 권한 관리


### chmod
파일 권한 변경


```python
# 실행 권한 추가
chmod +x script.sh

# 숫자로 권한 설정 (755: rwxr-xr-x)
chmod 755 file.txt

# 재귀적으로 권한 변경
chmod -R 755 mydir
```

### chown
파일 소유자 변경


```python
# 소유자 변경
sudo chown user file.txt

# 소유자와 그룹 변경
sudo chown user:group file.txt

# 재귀적으로 변경
sudo chown -R user:group mydir
```

## 프로세스 관리


### ps
실행 중인 프로세스 확인


```python
# 현재 터미널의 프로세스
ps

# 모든 프로세스 상세 정보
ps aux

# 특정 프로세스 검색
ps aux | grep python
```

### top
실시간 프로세스 모니터링


```python
top
# 종료: q
```

### htop
향상된 프로세스 모니터링 (설치 필요)


```python
htop
```

### kill
프로세스 종료


```python
# 프로세스 ID로 종료
kill 1234

# 강제 종료
kill -9 1234

# 프로세스 이름으로 종료
killall python
```

## 시스템 정보


### df

```python
# 사람이 읽기 쉬운 형식
df -h
```

### du

```python
# 현재 디렉토리 크기
du -sh .

# 하위 디렉토리별 크기
du -sh *

# 상위 10개 큰 디렉토리
du -h --max-depth=1 | sort -hr | head -10
```

## 네트워크


### ping
네트워크 연결 테스트


```python
# 4회 ping
ping -c 4 google.com
```

### curl
HTTP 요청 전송


```python
# GET 요청
curl https://api.example.com

# 헤더 포함 출력
curl -i https://api.example.com

# POST 요청
curl -X POST -d "data=value" https://api.example.com
```

### wget
파일 다운로드


```python
wget https://example.com/file.zip

# 백그라운드 다운로드
wget -b https://example.com/largefile.zip
```

### netstat

```python
# 모든 연결 확인
netstat -tuln

# 특정 포트 확인
netstat -tuln | grep 8000
```

### ss
소켓 통계 (netstat 대체)


```python
# 모든 TCP 연결
ss -tuln
```

## 패키지 관리


### apt
패키지 관리자


```python
# 패키지 목록 업데이트
sudo apt update

# 패키지 업그레이드
sudo apt upgrade

# 패키지 설치
sudo apt install package-name

# 패키지 제거
sudo apt remove package-name

# 패키지 검색
apt search keyword
```

## 압축 및 해제


### tar
압축 및 아카이브


```python
# 압축 (gzip)
tar -czvf archive.tar.gz folder/

# 압축 해제
tar -xzvf archive.tar.gz

# 압축 내용 확인
tar -tzvf archive.tar.gz
```

### zip/unzip
ZIP 압축


```python
# 압축
zip -r archive.zip folder/

# 압축 해제
unzip archive.zip
```

## 텍스트 처리


### echo
텍스트 출력


```python
echo "Hello World"

# 파일에 텍스트 쓰기 (덮어쓰기)
echo "text" > file.txt

# 파일에 텍스트 추가
echo "text" >> file.txt
```

### wc
줄, 단어, 문자 수 세기


```python
# 줄 수
wc -l file.txt

# 단어 수
wc -w file.txt

# 문자 수
wc -c file.txt
```

### sort
텍스트 정렬


```python
# 오름차순 정렬
sort file.txt

# 내림차순 정렬
sort -r file.txt
```

### uniq
중복 제거


```python
# 중복된 줄 제거 (정렬 필요)
sort file.txt | uniq

# 중복 횟수 카운트
sort file.txt | uniq -c
```
